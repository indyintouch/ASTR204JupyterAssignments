````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

#Read in the data file (there are many ways to do this in python, this is just one)
data = np.loadtxt("SN_m_tot_V2.0.txt",float)
x = data[:,2]
y = data[:,3]

#PLOTTING CODE here ##########
#Plot data. 'm-' tells python to plot a magenta coloured solid line.  
plt.plot(x, y, 'm-')
plt.title('Sunspot Number 1749-2019')
#This shows the limits of the plot
plt.ylim(-10,410)
plt.xlim(1749,2020)
#Label the axes
plt.xlabel("Year")
plt.ylabel("Sunspot Number")
###############################
````

It's really tricky to get a 11 year cycle to really line up, with the pattern. This whole 11 year thing is clearly an approximation, and the last cycle ("Cycle 24") has been strange, and is due to end any day. 

The below code attempts to overplot a sinusoid with an 11 year period over the data. Adjust the colors of the lines (and the zeropoint of the sinwave if you like) to make a verion you like. 

````python
#Plot 22 years of data with Poisson counting errors and overplot sine wave
plt.plot(x, y, 'm-',label='Sunspot data')

#y=sin x has a period of 2*pi, amplitude one and goes through the origin
#y=A+Bsin(2pi x/C + D) has amplitude B around y=A, has period C (in years) and starts at A,D
y2=80.+80*np.sin((2*np.pi/11.0)*(x-2011))
plt.plot(x,y2,'r-',label='Sine wave')
plt.title('Sunspot Number 1997-2019')
plt.legend()
plt.ylim(-10,270)
plt.xlim(1997,2020)
plt.xlabel("Year")
plt.ylabel("Sunspot Number")
plt.grid()
````

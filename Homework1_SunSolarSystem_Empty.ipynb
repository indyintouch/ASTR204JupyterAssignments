````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

#Read in the data file (there are many ways to do this in python, this is just one)
data = np.loadtxt("SN_m_tot_V2.0.txt",float)
x = data[:,2]
y = data[:,3]

#PLOTTING CODE here ##########
#Plot data. 'm-' tells python to plot a magenta coloured solid line.  
plt.plot(x, y, 'm-')
plt.title('Sunspot Number 1749-2021')
#This shows the limits of the plot
plt.ylim(-10,410)
plt.xlim(1749,2020)
#Label the axes
plt.xlabel("Year")
plt.ylabel("Sunspot Number")
###############################
````

3. *Will we see Sunspots this Semester?* Despite the name, sunspot number is not the number of spots on the Sun, as this obviously depends on the size of telescope and other factors (we'll hopefully explore this more in the Solar Observing Lab. Roughly the minimum sunspot number which creates sunspots visible in a small telescope (like our solar spotters) is $N=15$. 

Run the lines of code below, which do some simple maths and plot a second line, to add a line to the sunspot number data for the last five years. 

Notice how to comment out lines of code - you can play with adding a # and see what happens. Trial and error (and seeing error messages) are a normal part of coding. Try to play.

````python
#Plot data just in last 5 years and add visible subspot line (by removing the comment character "#" from the two lines)
plt.plot(x, y, 'm-',label='Sunspot data')
#Make a line at y=15 and overplot it.
z = 0*x + 15
plt.plot(x,z,'g',label='Rough limit for visible spots with small telescope')
plt.title('Sunspot Number 2016-2019')
plt.ylim(-10,100)
plt.xlim(2016,2020)
plt.legend()
plt.xlabel("Year")
plt.ylabel("Sunspot Number")
````

4. *Adding a sinusoidal curve and tweaking the function* 

It's really tricky to get a 11 year cycle to really line up, with the pattern. This whole 11 year thing is clearly an approximation, cycles recently have varied from 9-14 years in length, and the last cycle ("Cycle 24") was usually weak (https://earthsky.org/space/sunspot-cycle-25-among-strongest-on-record-says-ncar/)

The below code overplot a simple sinusoid with an 11 year period over the data. Adjust the colors and/or styles of the lines (and the zeropoint and/or period of the sinwave if you like) to make a version you think looks good. 

````python
#Plot 22 years of data with Poisson counting errors and overplot sine wave
plt.plot(x, y, 'm-',label='Sunspot data')

#y=sin x has a period of 2*pi, amplitude one and goes through the origin
#y=A+Bsin(2pi x/C + D) has amplitude B around y=A, has period C (in years) and starts at A,D
y2=80.+80*np.sin((2*np.pi/11.0)*(x-2011))
plt.plot(x,y2,'r-',label='Sine wave')
plt.title('Sunspot Number 1999-2021')
plt.legend()
plt.ylim(-10,270)
plt.xlim(1999,2022)
plt.xlabel("Year")
plt.ylabel("Sunspot Number")
plt.grid()
````

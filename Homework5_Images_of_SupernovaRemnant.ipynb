## Images in Python

Python makes it easy to include images in your plots, and even manipulate them. Let's play with an image of the Crab Nebula supernova remnant - but this code will work for any image you choose.  

This is a NASA image (available online here and other places: https://en.wikipedia.org/wiki/Crab_Nebula#/media/File:Crab_Nebula.jpg), and NASA provided the following caption: 

*This is a mosaic image, one of the largest ever taken by NASA's Hubble Space Telescope of the Crab Nebula, a six-light-year-wide expanding remnant of a star's supernova explosion. Japanese and Chinese astronomers recorded this violent event nearly 1,000 years ago in 1054, as did, almost certainly, Native Americans. The orange filaments are the tattered remains of the star and consist mostly of hydrogen. The rapidly spinning neutron star embedded in the center of the nebula is the dynamo powering the nebula's eerie interior bluish glow. The blue light comes from electrons whirling at nearly the speed of light around magnetic field lines from the neutron star. The neutron star, like a lighthouse, ejects twin beams of radiation that appear to pulse 30 times a second due to the neutron star's rotation. A neutron star is the crushed ultra-dense core of the exploded star. The Crab Nebula derived its name from its appearance in a drawing made by Irish astronomer Lord Rosse in 1844, using a 36-inch telescope. When viewed by Hubble, as well as by large ground-based telescopes such as the European Southern Observatory's Very Large Telescope, the Crab Nebula takes on a more detailed appearance that yields clues into the spectacular demise of a star, 6,500 light-years away. The newly composed image was assembled from 24 individual Wide Field and Planetary Camera 2 exposures taken in October 1999, January 2000, and December 2000. The colors in the image indicate the different elements that were expelled during the explosion. Blue in the filaments in the outer part of the nebula represents neutral oxygen, green is singly-ionized sulfur, and red indicates doubly-ionized oxygen.*

A colour image is a grid of numbers telling the computer how bright to make red, green and blue pixels. So we can read this in and plot it using ```matplotlib``` as below.

````python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

img = plt.imread('Crab_Nebula.jpg')
plt.imshow(img)
plt.show()
````

We can also explore just single layers of this image. For example:

````python
#Red image
plt.imshow(img[:,:,0], cmap="Reds_r")
plt.show()
````

````python
#Blue image
plt.imshow(img[:,:,2], cmap="Blues_r")
plt.show()
````

````python
#Green image
plt.imshow(img[:,:,1], cmap="Greens_r")
plt.show()
````

Notice that I have told python what colour to print those in - you don't have to plot the "Blue" data from the image as blue. It's held in the third layer: ```img[:,:,2]``` picks out the $i=2$ layer (python counts 0,1,2), and the ```:,:``` bit says use all the pixels in the image. I can plot the "blue" data in red if I like: 

(colour options are set by the ```cmap="XXXX"``` code with the allowed values of ```"XXXX"``` listed here: https://matplotlib.org/3.1.1/gallery/color/colormap_reference.html)

````python
#Blue image plotted red
plt.imshow(img[:,:,2], cmap="Reds_r")
plt.show()
````

# Multiplots

Here's a bit of code to make all the separate images side by side: 

````python
fig, ax = plt.subplots(1,4, figsize=(10,6))
ax[0].imshow(img[:,:,0], cmap="Reds_r")
ax[1].imshow(img[:,:,1], cmap="Greens_r")
ax[2].imshow(img[:,:,2], cmap="Blues_r")
ax[3].imshow(img);
for a in ax:
    a.set_xticklabels([])
    a.set_yticklabels([])
plt.show()
````

# Overplotting 

You can also put other plots commands into this in order to make them draw on top of your colour image. For example I can overplot the central pixel position as a big red "star" with the below command. For a full list of default plotting marker types see: https://matplotlib.org/3.1.0/api/markers_api.html#module-matplotlib.markers. Default colours here: https://matplotlib.org/2.0.2/api/colors_api.html (PS. I still have to look these up often). Hopefully ```markersize``` is fairly self explanatory - bigger number, bigger marker (PPS. but I had to look up that that's how you make the markers different sizes to write this). 

````python
plt.imshow(img)
[x,y]=[len(img)/2.0,len(img)/2.0]
plt.plot(x,y,'r*',markersize=12)
````
